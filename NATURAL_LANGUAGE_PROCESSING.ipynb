```python
all_uni_words = set()
all_uni_words.update(uni_words_one)
```

```python
all_uni_words.update(uni_words_two)
```

```python
all_uni_words
```

```python
full_vocab = dict()
i = 0
for word in all_uni_words:
  full_vocab[word] = i
  i = i + 1
```

```python
full_vocab
```

```python
one_freq = [0]*len(full_vocab)
two_freq = [0]*len(full_vocab)
all_words = ['']*len(full_vocab)
```
